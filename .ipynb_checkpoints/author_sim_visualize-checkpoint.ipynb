```python
from keras.preprocessing.text import one_hot, Tokenizer
from keras.preprocessing.sequence import pad_sequences
from keras.models import Sequential
from keras.layers import Dense, Input, GlobalAveragePooling1D
from keras.layers import Flatten, Dot
from keras.layers.embeddings import Embedding
from keras.models import Model
from umap import UMAP
import numpy as np
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
import pickle

import warnings
warnings.filterwarnings("ignore")
```

```python
def emb_layer(input_tens, embedding_matrix, input_length, name):
	x = Embedding(
		vocab_size, 
		100, 
		input_length=input_length, 
		weights=[embedding_matrix],
		trainable=False
	)(input_tens)
	x = GlobalAveragePooling1D()(x)
	#x = Flatten()(x)
	x = Dense(100, activation='linear', name=name)(x)

	return x
```

```python
padded_docs = np.load('padded_docs.npy')
with open('labels', 'rb') as f:
    padded_labels = pickle.load(f)
    padded_labels = np.array(padded_labels)
pair_labels = np.load('pair_labels.npy')
pair_idx = np.load('pairs_idx.npy')

embedding_matrix = np.load('glov_emb_matrix.npy')

print('Padded Docs:', padded_docs.shape)
print('Padded Labels:', padded_labels.shape)
print('Pair Labels:', pair_labels.shape)
print('Pair Idx Matches:', pair_idx.shape)
print('Embedding Matrix:', embedding_matrix.shape)

input_length = 10000
vocab_size = 544698
```

```python
input_tens_1 = Input(shape=(input_length,), name='input_1')
input_tens_2 = Input(shape=(input_length,), name='input_2')
emb_1 = emb_layer(input_tens_1, embedding_matrix, input_length, 'emb_1')
emb_2 = emb_layer(input_tens_2, embedding_matrix, input_length, 'emb_2')
merged = Dot(normalize=True, axes=1)([emb_1, emb_2])
out = Dense(1, activation='sigmoid')(merged)

base_model = Model(inputs=[input_tens_1, input_tens_2], outputs=out)
base_model.load_weights('ckpt_weights/model_weights_e_32')
base_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

base_model.summary()
```

```python
model = Model(inputs=base_model.get_layer('input_1').input, outputs=base_model.get_layer('emb_1').output)
model.summary()
```

```python
embeddings = model.predict(padded_docs, verbose=1)
print('Embeddings:', embeddings.shape)
```

```python
umap = UMAP(n_neighbors=20, metric='cosine')
vis_embeddings = umap.fit_transform(embeddings)
print('Vis Embeddings:', vis_embeddings.shape)
```

```python
author_df = pd.read_csv('author_label_df.csv')
author_df.head(10)
```

```python
fig = plt.figure(figsize=(15,8))
uniq_labels = np.unique(padded_labels)
for label in uniq_labels:
    author = author_df.loc[author_df['label'] == label, 'author'].values[0]
    label_idx = np.argwhere(padded_labels == label)
    sub_vis_embeddings = vis_embeddings[label_idx, :]
    sub_vis_embeddings = sub_vis_embeddings.reshape(
        sub_vis_embeddings.shape[0]*sub_vis_embeddings.shape[1],
        sub_vis_embeddings.shape[2]
    )
    plt.scatter(
        sub_vis_embeddings[:, 0], 
        sub_vis_embeddings[:, 1], 
        c=np.random.rand(1,3), 
        label=str(author), 
        alpha=0.4
    )

plt.title('Visualized Embeddings for Works, Colored by Author')
plt.xlabel('UMAP 1')
plt.ylabel('UMAP 2')
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.10),
          fancybox=True, shadow=True, ncol=5)
plt.show()
```
